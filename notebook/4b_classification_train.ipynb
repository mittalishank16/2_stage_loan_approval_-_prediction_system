```python
# train set
X_train = train_original.drop(columns='loan_approval', axis = 1)
y_train = train_original['loan_approval']

# test set
X_test = test_original.drop(columns='loan_approval', axis = 1)
y_test = test_original['loan_approval']
```

```python
Classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Ridge Classifier': RidgeClassifier(random_state=42),
    'KNN Classifier': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(eval_metric='logloss', random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'Extra Trees': ExtraTreesClassifier(random_state=42),
    'SVM Classifier': SVC(probability=True, random_state=42),
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(random_state=42, max_iter=1000)
}
```

```python
def fit_score_model(model, model_name, X_train_prep, y_train_prep):
    # Define the directory and the file path
    model_dir = Path('saved_models') / model_name
    accuracy_scores_path = model_dir / f'accuracy_{model_name}.pkl'
    
    # 1. Ensure the directory exists (this prevents the FileNotFoundError on dump)
    model_dir.mkdir(parents=True, exist_ok=True)
    
    # 2. Check if file exists using .exists() instead of .resolve(strict=True)
    if accuracy_scores_path.exists():
        accuracy_scores = joblib.load(accuracy_scores_path)
        print(f"--- Loaded cached scores for {model_name} ---")
    else:
        print(f"--- Training and Cross-Validating {model_name} ---")
        # Get Accuracy Scores via Cross-Validation
        accuracy_scores = cross_val_score(model, X_train_prep, y_train_prep, cv=3, scoring='accuracy')
        # Save the scores
        joblib.dump(accuracy_scores, accuracy_scores_path)

    # Print accuracy metrics
    print('Accuracy Scores: {}'.format(accuracy_scores))
    print('Mean Accuracy: {:.4f}'.format(accuracy_scores.mean()))
    print('Standard deviation: {:.4f}'.format(accuracy_scores.std()))

    # 3. Generate predictions for Confusion Matrix and Classification Report
    y_pred = cross_val_predict(model, X_train_prep, y_train_prep, cv=3)

    print('\n--- Confusion Matrix ---')
    print(confusion_matrix(y_train_prep, y_pred))

    print('\n--- Classification Report ---')
    print(classification_report(y_train_prep, y_pred))
```

```python
def folder_check_model(model_name):
    if not os.path.exists('saved_models_classification/{}/'.format(model_name)):
        os.makedirs('saved_models_classification/{}/'.format(model_name))
```

```python
# Loop over all classifiers
for model_name, model in Classifiers.items():
    print('\n')
    print('  {}  '.center(50,'-').format(model_name))
    
    folder_check_model(model_name)
    fit_score_model(model, model_name, X_train, y_train)
```

```python
# using neural network
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

# Defining the architecture
model = Sequential([
    # Input Layer + First Hidden Layer
    Dense(128, input_shape=(X_train.shape[1],), activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    # Second Hidden Layer
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    # Third Hidden Layer
    Dense(32, activation='relu'),
    Dropout(0.2),

    # Fourth Hidden Layer
    Dense(16, activation='relu'),
    
    # Output Layer (Sigmoid for binary classification: 0 to 1 probability)
    Dense(1, activation='sigmoid')
])

# Compiling the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Setting up Early Stopping to prevent overfitting
early_stopper = EarlyStopping(
    monitor='val_loss', 
    patience=15, 
    restore_best_weights=True
)

# Training the model
history = model.fit(
    X_train, y_train, 
    epochs=200, 
    batch_size=32, 
    validation_split=0.2, # Uses 20% of data to check for overfitting during training
    callbacks=[early_stopper],
    verbose=1
)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def evaluate_keras_model(model, X_test, y_test, model_name="Deep Neural Network"):
    print(f'  {model_name} Evaluation  '.center(50, '-'))
    
    # 1. Get probabilities and convert to binary classes (0 or 1)
    y_probs = model.predict(X_test)
    y_pred = (y_probs > 0.5).astype(int).flatten()
    
    # 2. Calculate Accuracy
    acc = accuracy_score(y_test, y_pred)
    
    print(f'Accuracy: {acc:.4f}')
    
    # 3. Generate Confusion Matrix
    print('\n--- Confusion Matrix ---')
    print(confusion_matrix(y_test, y_pred))
    
    # 4. Generate Classification Report
    print('\n--- Classification Report ---')
    print(classification_report(y_test, y_pred))

# Execute the evaluation
evaluate_keras_model(model, X_test, y_test)
```
